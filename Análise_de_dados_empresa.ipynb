```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount ("/content/gdrive")
```

```python
planilha = pd.read_excel("/content/gdrive/MyDrive/Projeto /controle_da_empresa.xlsx",None)
df = planilha["Vendas"]
df.head()
```

```python
plt.hist("Total de Vendas", data=df)
plt.title("Total de vendas")
plt.xlabel("Quantidade de vendas")
plt.ylabel("Ocorrências")
plt.show()
```

```python
v1 = df["Data da Venda"]
v2 = df["Total de Vendas"]
df["Data da Venda"] = pd.to_datetime(df["Data da Venda"], infer_datetime_format=True)
plt.plot(v1,v2)
plt.ylabel("Total em Vendas")
plt.xlabel("Data")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar("Vendedor", "Total de Vendas", data=df)
plt.title("Vendas por vendedor")
plt.xlabel("Vendedor")
plt.ylabel("Quantidade de vendas")
plt.show()
```

```python
a = dict(df["Item"].value_counts())
labels = a.keys()
valores = a.values()
plt.pie(valores, labels = labels)
plt.title("Produtos")
plt.show()
```

```python
df.groupby('Vendedor')['Total de Vendas'].sum().plot.bar();
```
